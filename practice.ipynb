```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
# Coin Toss
# # Define probability of success
p = 0.5
x = [0,1] # unique outcomes

# Get the probability mass function
pmf = bernoulli.pmf(x,p)

# Plot the PMF
plt.figure()
plt.bar(x, pmf, tick_label=["Failure (0)", "Success (1)"], color=['red', 'green'])
plt.title(f"Bernoulli Distribution PMF (p={p})")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Ad Click
# probability of clicking on an ad
p = 0.3
click_prob = bernoulli.pmf([1],p)
no_click_prob = bernoulli.pmf([0],p)
print(f"Probabilty of clicking on an ad: {click_prob[0]:.2f}")
print(f"Probabilty of not clicking on an ad: {no_click_prob[0]:.2f}")
```

```python
# binomial distribution
from scipy.stats import binom

# Define parameters
n = 10  # number of trials
p = 0.5  # probability of success
x = np.arange(0, n+1)  # possible number of successes

# Get the probability mass function
pmf = binom.pmf(x, n, p)

# Plot the PMF
plt.figure()
plt.bar(x, pmf, tick_label=x, color='skyblue', alpha=0.7)
plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# basketball free throws

# Define parameters
n = 10  # number of trials
p = 0.80
x = 8  # number of successes
# Get the probability of making exactly 8 free throws
pmf = binom.pmf(x, n, p)
print(f"Probability of making exactly {x} free throws out of {n}: {pmf:.4f}")
# Get the probability of making at least 8 free throws
# includes 8, 9, and 10 successes
pmf_at_least_8 = binom.pmf(x, n, p) + binom.pmf(x+1, n, p) + binom.pmf(x+2, n, p)
print(f"Binom: Probability of making at least {x} free throws out of {n}: {pmf_at_least_8:.4f}")
# or using the cumulative distribution function (CDF)
# 1 - probability of making at most 7 free throws
pmf_at_least_8 = 1 - binom.cdf(x-1, n, p)
print(f"CDF: Probability of making at least {x} free throws out of {n}: {pmf_at_least_8:.4f}")
```

#### Exponential Distributions

```python
from scipy.stats import expon

# parameters
lambda_ = 0.1 # time between events per day
scale = 1/lambda_
x = np.linspace(0, 50, 1000) # list of times until next event
pdf = expon.pdf(x, scale=scale)
# plot the PDF
plt.figure()
plt.plot(x, pdf, color='blue')
plt.title(f"Exponential Distribution PDF (λ={lambda_})")
plt.xlabel("Time Until Next Event")
plt.ylabel("Probability Density")
plt.grid()
plt.axvline(scale, color='red', linestyle='--', label='Mean')
plt.axvline(scale*2, color='green', linestyle='--', label='2x Mean')
plt.axvline(scale*3, color='orange', linestyle='--', label='3x Mean')
plt.axvline(scale*4, color='purple', linestyle='--', label='4x Mean')
plt.legend()
plt.show()
```

### Beta Distribution

```python
from scipy.stats import beta

# parameters
a, b = 2, 5 # shape parameters
x = np.linspace(0, 1, 1000) # list of probabilities
pdf = beta.pdf(x, a, b)
# plot the PDF
plt.figure()
plt.plot(x, pdf, color='blue')
plt.title(f"Beta Distribution PDF (α={a}, β={b})")
plt.xlabel("Probability")
plt.ylabel("Probability Density")
plt.grid()
# find the mean
mean = a / (a + b)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.show()

# Interpretation: the x-axis shows the probability of success, 
# while the y-axis shows the probability of that probability of success occurring.
# useful for studying models with have two outcomes (success/failure)
# and we don't know the probability of success a priori, 
# so we assume a uniform distribution of probabilities
# and then update our beliefs based on observed data.
```

```python
# Example: Bayesian A/B Testing

# Version A: 100 users, 40 cicks

# CTR: click-through rate
# CTR = clicks / total_users

a = 40 + 1 # successes + 1
b = 100 - 40 + 1 # failures + 1
x = np.linspace(0, 1, 1000) # list of probabilities
pdf_a = beta.pdf(x, a, b)
# plot the PDF
plt.figure()
plt.plot(x, pdf_a, color='blue', label='Version A')
plt.title(f"Beta Distribution PDF (α={a}, β={b})")
plt.xlabel("Probability")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

# Version B: 100 users, 30 clicks
a = 30 + 1 # successes + 1
b = 100 - 30 + 1 # failures + 1
x = np.linspace(0, 1, 1000) # list of probabilities
pdf_a = beta.pdf(x, a, b)
# plot the PDF
plt.figure()
plt.plot(x, pdf_a, color='blue', label='Version B')
plt.title(f"Beta Distribution PDF (α={a}, β={b})")
plt.xlabel("Probability")
plt.ylabel("Probability Density")
plt.grid()
plt.show()

# What is the probability that B has a higher CTR than A?
# i.e. P(P_A(x) > P_A(x)) 
# To do this we can use 1000 samples
samples_A = np.random.beta(40+1, 100-40+1, size=100_000)
samples_B = np.random.beta(50+1, 100-50+1, size=100_000)
print(samples_A[:10])
# count the number of times B is greater than A
n_wins = np.sum(samples_B > samples_A)
# calculate the probability
probability = np.mean(samples_B > samples_A)
print(f"Number of times that B has a higher CTR than A: {n_wins}")
print(f"Probability that B has a higher CTR than A: {probability:.4f}")
```

```python
# code from MD file
# Observed data
clicks_A, views_A = 40, 100
clicks_B, views_B = 50, 100

# Prior parameters (uniform prior)
alpha_prior, beta_prior = 1, 1

# Posterior parameters (these are determined through Bayesian inference. For now assume they are given)
alpha_A = alpha_prior + clicks_A
beta_A = beta_prior + views_A - clicks_A
alpha_B = alpha_prior + clicks_B
beta_B = beta_prior + views_B - clicks_B

# Sample from posteriors (you will learn sampling below. It means generate random data based on that distribution)
samples = 100_000
posterior_A = np.random.beta(alpha_A, beta_A, samples)
posterior_B = np.random.beta(alpha_B, beta_B, samples)

# Probability that B > A
prob_B_beats_A = np.mean(posterior_B > posterior_A)
print(f"Probability that version B has a higher CTR than version A: {prob_B_beats_A:.4f}")

# Plotting
x = np.linspace(0, 1, 1000)
plt.figure(figsize=(10, 6))
plt.plot(x, beta.pdf(x, alpha_A, beta_A), label='Posterior A', color='blue')
plt.plot(x, beta.pdf(x, alpha_B, beta_B), label='Posterior B', color='green')
plt.title('Posterior Distributions of Click-through Rates')
plt.xlabel('CTR')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

### Oversampling


```python
# Random Oversampling

from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.datasets import make_classification

X, y = make_classification(n_classes=2, weights=[0.9,0.1], n_samples=1000, random_state=42)
n_0 = sum(y == 0)
n_1 = sum(y == 1)
print(f"Samples of class 0: {n_0}")
print(f"Samples of class 1: {n_1}")
print(f"Original dataset shape: {Counter(y)}")

# Draw the original dataset
plt.figure(figsize=(10, 5))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Class 0', alpha=0.5)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Class 1', alpha=0.5)
print(f"X shape: {X.shape}")
plt.title('Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()

# Resample the dataset
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)
n_0 = sum(y_resampled == 0)
n_1 = sum(y_resampled == 1)
print(f"Samples of class 0: {n_0}") 
print(f"Samples of class 1: {n_1}")
print(f"Resampled dataset shape: {Counter(y_resampled)}")

# Draw the resampled dataset
plt.figure(figsize=(10, 5))
print(f"X_resampled shape: {X_resampled.shape}")
plt.scatter(X_resampled[y_resampled == 0][:,0], X_resampled[y_resampled == 0][:,1], label='Class 0', alpha=0.5)
plt.scatter(X_resampled[y_resampled == 1][:,0], X_resampled[y_resampled == 1][:,1], label='Class 1', alpha=0.5)
print(f"X_resampled shape: {X_resampled.shape}")
plt.title('Resampled Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```

```python
# synthetic minority oversampling technique
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X,y)
print(f"After SMOTE, samples of class 0: {sum(y_smote == 0)}")
print(f"After SMOTE, samples of class 1: {sum(y_smote == 1)}")
print(f"After SMOTE, resampled dataset shape: {Counter(y_smote)}")
```

```python
# Tomek Links
from imblearn.under_sampling import TomekLinks
tl = TomekLinks()
X_tl, y_tl = tl.fit_resample(X,y)
print(f"After Tomek Links: samples of class 0: {sum(y_tl == 0)}")
print(f"After Tomek Links: samples of class 1: {sum(y_tl == 1)}")
print(f"After Tomek Links: resampled dataset shape: {Counter(y_tl)}")

# Draw the resampled dataset
plt.figure(figsize=(10, 5))
plt.scatter(X_tl[y_tl==0][:,0], X_tl[y_tl==0][:,1], label='Class 0', alpha=0.5)
plt.scatter(X_tl[y_tl==1][:,0], X_tl[y_tl==1][:,1], label='Class 1', alpha=0.5)
plt.title('Resampled Dataset (Random Undersampling)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```

```python
# combine oversampling and undersampling
from imblearn.combine import SMOTETomek

smote_tomek = SMOTETomek(random_state=42)
X_combined, y_combined = smote_tomek.fit_resample(X,y)
print(f"After SMOTE + Tomek:", Counter(y_combined))
print(f"After SMOTE + Tomek: samples of class 0: {sum(y_combined == 0)}")
print(f"After SMOTE + Tomek: samples of class 1: {sum(y_combined == 1)}")

# Draw the resampled dataset
plt.figure(figsize=(10, 5))
plt.scatter(X_combined[y_combined==0][:,0], X_combined[y_combined==0][:,1], label='Class 0', alpha=0.5) 
plt.scatter(X_combined[y_combined==1][:,0], X_combined[y_combined==1][:,1], label='Class 1', alpha=0.5)
plt.title('Resampled Dataset (SMOTE + Tomek)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Load dataset
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Standardize features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Initialize
remaining_features = list(X_scaled.columns)
selected_features = []
best_score = 0
max_features = len(X_scaled.columns)  # Or set a custom cap

while remaining_features and len(selected_features) < max_features:
    scores = []
    for feature in remaining_features:
        current_features = selected_features + [feature]
        #print(f"current_features", current_features)
        X_subset = X_scaled[current_features]
        clf = LogisticRegression(max_iter=1000, solver='liblinear')
        score = cross_val_score(clf, X_subset, y, cv=5).mean()
        scores.append((feature, score))

    # Pick the best new feature
    best_feature, best_feature_score = max(scores, key=lambda x: x[1])

    # If it improves performance, add it
    if best_feature_score > best_score:
        selected_features.append(best_feature)
        remaining_features.remove(best_feature)
        best_score = best_feature_score
        print(f"✅ Added: {best_feature} (CV Accuracy = {best_feature_score:.4f})")
    else:
        print(f"⛔ No improvement by adding more features.")
        break

print("\n🏁 Final selected features:")
print(selected_features)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

# Load dataset
data = load_wine()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

# Standardize features
scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

# Initialize
remaining_features = list(X_scaled.columns)
removed_features = []
best_score = 0
max_removed_features = len(X_scaled.columns)  # Or set a custom cap

while remaining_features and len(removed_features) < max_removed_features:
    scores = []
    for feature in remaining_features:
        condition = lambda x: x == feature 
        current_features = [x for x in X_scaled.columns if not condition(x)]
        #print(f"current_features", current_features)
        X_subset = X_scaled[current_features]
        clf = LogisticRegression(max_iter=1000, solver='liblinear')
        score = cross_val_score(clf, X_subset, y, cv=5).mean()
        scores.append((feature, score))
        print(f"When removing {feature}, we get a score {score:3f}")

    # Pick the best new feature
    best_feature, best_feature_score = max(scores, key=lambda x: x[1])

    # If it improves performance, add it
    if best_feature_score > best_score:
        removed_features.append(best_feature)
        remaining_features.remove(best_feature)
        best_score = best_feature_score
        print(f"✅ Removed: {best_feature} (CV Accuracy = {best_feature_score:.4f})")
    else:
        print(f"⛔ No improvement by removing more features.")
        break

print("\n🏁 Final selected features:")
print(selected_features)
```
